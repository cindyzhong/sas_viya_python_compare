# Data Mining Workflow - Python

```python
# Import necessary packages and modules
import pandas as pd
import collections
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Exploration and Preparation 

## Investigate the data - Python

```python
# Reading the SAS Dataset using Python
df =  pd.read_sas("hmeq_train.sas7bdat")

# Print out first 10 rows of the dataset
df.head(10)
```

```python
# Look at the shape
df.shape
```

# Gather Summary Statistics

```python
# Create table of summary statistics
df.describe(include = "all")
```

## Plot Numeric Variable Distribution - Python

```python
df.hist(figsize = (15,10))
```

# Perform Imputation on Missing Values - Python

```python
# Replace the numeric columns with missing value to be the median
df_imp = df.fillna (df.median())
df_imp.head(10)
```

```python
# Replace the categorical columns with missing value to be the mode
df_imp = df_imp.fillna(df_imp.mode().iloc[0])
df_imp.head(10)
```

## Encoding Categorical Variables - Python

```python
# sklearn do not take categorical variables directly, need to make into dummy variables
df_clean = pd.get_dummies(df_imp, prefix=['REASON', 'JOB'])
df_clean.columns
df_clean[:10]
```

# Partition Data into Training and Validation Sets

```python
# Our target variable is BAD, and the rest are our feature set
target = df_clean.values[:,0]
feature = df_clean.values[:,1::]

# Create a 70/30 simple random sample split
from sklearn.cross_validation import train_test_split
target_train, target_val, feature_train, feature_val = train_test_split( target, feature, test_size = 0.3, random_state = 100)
```

# Build Models

## Decision Tree - Python

```python
from sklearn import tree
# leaving everything as default, like the CAS version
ds = tree.DecisionTreeClassifier(criterion='entropy')
ds = ds.fit(feature_train,target_train)
```

## Gradient Boosting Tree - Python

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb = gb.fit(feature_train,target_train)
```

## Neural Network - Python

```python
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier()
nn = nn.fit(feature_train,target_train)
```

## SVM - Python

```python
from sklearn import svm
svm_clf = svm.SVC()
svm_clf = svm_clf.fit(feature_train,target_train)
```

# Assess Performance - Python

```python
from sklearn import metrics
def measure_performance(X,y,clf, show_accuracy=True, show_misclassification=False,show_classification_report=False, show_confusion_matrix=False):
    y_pred=clf.predict(X)   
    if show_accuracy:
        print ("Accuracy:{0:.3f}".format(metrics.accuracy_score(y,y_pred)))

    if show_misclassification:
        print ("Misclassification:{0:.3f}".format(1 - metrics.accuracy_score(y,y_pred)))
        
    if show_classification_report:
        print ("Classification report")
        print (metrics.classification_report(y,y_pred))
        
    if show_confusion_matrix:
        print ("Confusion matrix")
        print (metrics.confusion_matrix(y,y_pred))
```

```python
print "Decision Tree: " 
measure_performance(feature_val,target_val,ds,show_accuracy=False, show_misclassification=True)
print "Gradient Boost: " 
measure_performance(feature_val,target_val,gb,show_accuracy=False, show_misclassification=True)
print "Neural Network" 
measure_performance(feature_val,target_val,nn,show_accuracy=False, show_misclassification=True)
print "Neural Network" 
measure_performance(feature_val,target_val,svm_clf,show_accuracy=False, show_misclassification=True)
```
